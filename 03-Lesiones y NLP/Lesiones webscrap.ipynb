```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from tabulate import tabulate
```

Optamos por hacer scrapping con selenium, para obtener los lesionados de La Liga de la página futbolfantasy.com

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.futbolfantasy.com/laliga/lesionados"
driver.get(url)

time.sleep(5)

jugadores = driver.find_elements(By.CLASS_NAME, "jugador")

jugadores_lesionados = [jugador.text.strip() for jugador in jugadores]
print("Jugadores lesionados:", jugadores_lesionados)

driver.quit()
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.futbolfantasy.com/laliga/lesionados"
driver.get(url)

time.sleep(5)

jugadores = driver.find_elements(By.CLASS_NAME, "jugador")
equipos = driver.find_elements(By.CLASS_NAME, "equipo")
lesiones = driver.find_elements(By.CLASS_NAME, "lesion")

data = []
for i in range(len(jugadores)):
    jugador = jugadores[i].text.strip()
    equipo = equipos[i].text.strip() if i < len(equipos) else "Desconocido"
    lesion = lesiones[i].text.strip() if i < len(lesiones) else "No especificado"
    data.append({"Jugador": jugador, "Equipo": equipo, "Lesión": lesion})

df = pd.DataFrame(data)

driver.quit()

df
```

Tenemos que afinar el scrapping con las clases y el codigo html para seleccionar correctamente los equipos y la lesión de cada jugador

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.futbolfantasy.com/laliga/lesionados"
driver.get(url)

time.sleep(5)

jugadores = driver.find_elements(By.CSS_SELECTOR, ".jugador_principal .block-new .datos a, .equipo .block-new .datos a, .lesionados .block-new .datos a")
equipos = driver.find_elements(By.CSS_SELECTOR, ".lesionados .block-new header.title")
lesiones = driver.find_elements(By.CSS_SELECTOR, ".equipo .block-new .datos div, .lesionados .block-new .datos div")

data = []
for i in range(len(jugadores)):
    jugador = jugadores[i].text.strip()
    equipo = equipos[i].text.strip() if i < len(equipos) else "Desconocido"
    lesion = lesiones[i].text.strip() if i < len(lesiones) else "No especificada"
    data.append({"Jugador": jugador, "Equipo": equipo, "Lesión": lesion})


df_lesiones = pd.DataFrame(data)

print(tabulate(df_lesiones, headers='keys', tablefmt='grid'))

driver.quit()
```

Tenemos algunos equipos que se nos siguen marcando como 'Desconocido', asi que vamos a retocar el código de nuevo

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.futbolfantasy.com/laliga/lesionados"
driver.get(url)

time.sleep(5)

equipos_banners = driver.find_elements(By.CSS_SELECTOR, ".block-new header.title")

jugadores = driver.find_elements(By.CSS_SELECTOR, ".jugador_principal .block-new .datos a, .equipo .block-new .datos a, .lesionados .block-new .datos a")
lesiones = driver.find_elements(By.CSS_SELECTOR, ".equipo .block-new .datos div, .lesionados .block-new .datos div")


data = []
current_team = None

for i in range(len(jugadores)):
    if i < len(equipos_banners):
        current_team = equipos_banners[i].text.strip()

    jugador = jugadores[i].text.strip()
    lesion = lesiones[i].text.strip() if i < len(lesiones) else "No especificado"
    equipo = current_team if current_team else "Desconocido"
    data.append({"Jugador": jugador, "Equipo": equipo, "Lesión": lesion})

df_lesiones_laliga = pd.DataFrame(data)

print(tabulate(df_lesiones_laliga, headers='keys', tablefmt='grid'))

driver.quit()
```

Vamos a crear una columna nueva para el tiempo de la baja, porque esta información nos la scrappea en conjunto con el tipo de lesión

```python
df_lesiones_laliga[['Lesión', 'Tiempo de baja']] = df_lesiones_laliga['Lesión'].str.split('\n', expand=True)

print(tabulate(df_lesiones_laliga, headers='keys', tablefmt='grid'))
```
